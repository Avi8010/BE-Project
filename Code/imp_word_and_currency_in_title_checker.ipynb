```python
print(len(important_words))
```

```python
def count_target_words(text):
    words = str(text).lower().split()
    return sum(word in important_words for word in words)
```

```python
df.columns
```

```python
df['Important_Word_Count'] = df['description'].apply(count_target_words)
```

```python
df.head(10)
```

```python
def mark_important_words(num):
  if(num>0):
    return 1
  else:
    return 0
```

```python
df['important_words'] = df['Important_Word_Count'].apply(mark_important_words)
```

```python
df.head()
```

```python
currency = ["$", "€", "¥", "£", "₹"]
```

```python
def currency_in_title(text):
  words = str(text).lower().split()
  for word in words:
    for j in range(len(word)):
      if(word[j] in currency):
        return 0
  return 1
```

```python
df['currency_in_title'] = df['job_title'].apply(currency_in_title)
```

```python
cnt = 0
for index, row in df.iterrows():
  if(row['currency_in_title']==0):
    cnt+=1
print(cnt)
```
