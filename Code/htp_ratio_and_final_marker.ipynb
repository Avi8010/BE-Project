```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('internships.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
def has_htp_ratio(row):
  if(row['hired_to_posted_ratio']>=0.5 and row['hired_to_posted_ratio']<=7):
    return 1
  return 0
```

```python
df['htp_ratio'] = df.apply(has_htp_ratio, axis=1)
```

```python
def is_authentic(row):
    if(row['company_description'] and row['has_company_logo'] and row['job_description_length'] and row['important_words'] and row['currency_in_title'] and row['htp_ratio']):
      return 1
    return 0
```

```python
df['authentic'] = df.apply(is_authentic, axis=1)
```
